```python
import folium
import s2sphere
```

```python
sw_corner = s2sphere.LatLng.from_degrees(40.615643, -74.038159)
nw_corner = s2sphere.LatLng.from_degrees(40.894218, -73.841595)
nyc_rect = s2sphere.LatLngRect.from_point_pair(sw_corner, nw_corner)
nyc_coverer = s2sphere.RegionCoverer()
# 17 is small enpough to divde blocks, I wouldn't anticipate any two stations being in the same parcel.
# # There may be overlap with 16, but they'd be around a block or two apart at most.
nyc_coverer.min_level = 16
nyc_coverer.max_level = 16
nyc_coverer.max_cells = 150000
nyc_parcels = nyc_coverer.get_covering(nyc_rect)
```

```python
nyc_map = folium.Map(location = [40.766858, -73.977855], zoom_start = 13)

for cell_id in nyc_parcels:
    parcel = s2sphere.Cell(cell_id)
    vetices = []
    for i in range(4):
        vertex = parcel.get_vertex(i)
        latlng = s2sphere.LatLng.from_point(vertex)
        vetices.append((latlng.lat().degrees, latlng.lng().degrees))
    
    folium.Polygon(locations = vetices, color = 'navy', opacity=0.1, fill = False).add_to(nyc_map)

nyc_map.save("../Output/nyc_map.html")
```
